```python
def add_enhanced_features(RE):
    RE['PricePerSF'] = RE['SalePrice'] / RE['GrLivArea']
    RE['PropertyAge'] = RE['YrSold'] - RE['YearBuilt']
    RE['YearsSinceRemodel'] = RE['YrSold'] - RE['YearRemodAdd']
    RE['TotalPorchSF'] = RE['OpenPorchSF'] + RE['EnclosedPorch'] + RE['3SsnPorch'] + RE['ScreenPorch']
    RE['TotalOutsideSF'] = RE['GarageArea'] + RE['TotalPorchSF']
    RE['TotalSF'] = RE['TotalBsmtSF'] + RE['1stFlrSF'] + RE.get('2ndFlrSF', 0)
    RE['IsRemodeled'] = (RE['YearBuilt'] != RE['YearRemodAdd']).astype(int)
    RE['TotalBathrooms'] = (
    RE['FullBath'] + 0.5 * RE['HalfBath'] +
    RE.get('BsmtFullBath', 0) + 0.5 * RE.get('BsmtHalfBath', 0)
    )
    RE['HasGarage'] = RE['GarageArea'].apply(lambda x: 1 if x > 0 else 0)
    RE['HasBsmt'] = RE['TotalBsmtSF'].apply(lambda x: 1 if x > 0 else 0)
    RE['HasFireplace'] = RE['Fireplaces'].apply(lambda x: 1 if x > 0 else 0)
    RE['HasPorch'] = (RE.get('OpenPorchSF', 0) + RE.get('EnclosedPorch', 0) +
                      RE.get('3SsnPorch', 0) + RE.get('ScreenPorch', 0)).apply(lambda x: 1 if x > 0 else 0)
    return df
```

- Feature Engineering- Creation of new variables from existing variables.
- PricePerSF- It is the price of the property per square foot.
- PropertyAge- It is the age of property at the time of it being sold.
- YearsSinceRemodel- It is the time in years since the property is remodeled.
- TotalPorchSF- It is the total porch area of the property in square foot.
- TotalOutsideSF- It is the total outside area of the property(Excluding living area).
- TotalSF- It i the total living area of the property.
- IsRemodeled- It tells if the property is remodeled after being built or not(1= Is remodeled, 0= Is not remodeled).
- TotalBathroom- Total number of bathrooms on the property.
- HasGarage- It tells if the property has garage or not(1= Has garage, 0= Does not have garage).
- HasBsmt- It tells if the property has basement or not(1= Has basement, 0= Does not have basement).
- HasFireplace- It tells if the property has fireplaces or not(1= Has fireplace, 0= Does not have fireplace).
- HasPorch- It tells if the property has porch or not(1= Has porch, 0= Does not have porch).
